```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cantilever(w=2.5, t=1.5, L=100.0, X=300.0, Y=600.0, 
              T=0, E0=29778106.1, Es=-4999.82107):
    E = E0 + Es*T
    d = 4*L**3/(E*w*t)*np.sqrt((Y/t**2.0)**2.0 + (X/w**2.0)**2.0)
    return d
```

```python
T = np.linspace(-20,500, 20)
np.random.seed(1234)
d400 = cantilever(T=T, X=0, Y=400)
d400n1 = d400 + np.random.normal(loc=0, scale=0.1,size=(20,))
d400n2 = d400 + np.random.normal(loc=0, scale=0.15,size=(20,))
d600 = cantilever(T=T, X=0, Y=600)
d600n = d600 + np.random.normal(loc=0, scale=0.15,size=(20,))
plt.plot(T,d400n1,".", T,d600n,".",T,d400, T,d600)
```

```python
plt.plot(T,d400n1,".", label="400 lb")
plt.plot(T,d600n,".", label="600 lb")
plt.xlabel("Temperature (F)")
plt.ylabel("Displacement (in)")
plt.legend()
#plt.savefig("displacements_two_loads.png")
```

```python
E0 = 2.9746942319e+07
Es = -4.8549276396e+03
plt.plot(T,d400n1,".", T, cantilever(T=T, X=0, Y=400, E0=E0, Es=Es))
plt.xlabel("Temperature (F)")
plt.ylabel("Displacement @ 400 lb (in)")
#plt.savefig("residualresult.png")
```

```python
E0 = 2.9792117040e+07
Es = -5.2921595180e+03
plt.plot(T,d400n1,".", label="Experiment 1")
plt.plot(T, d400n2,".", label="Expermient 2")
plt.plot(T, cantilever(T=T, X=0, Y=400, E0=E0, Es=Es), label="Best fit")
plt.legend()
plt.xlabel("Temperature (F)")
plt.ylabel("Displacement @ 400 lb (in)")
#plt.savefig("replicateresult.png")
```
